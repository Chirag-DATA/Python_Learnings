```python
import matplotlib.pyplot as plt
```

#### LINE CHARTS

```python
x = ["Mon","Tue","Wed","Thu","Fri","Sat"]
y = [10,5,12,7,20,15]

plt.plot(x,y)             #function to plot the graph 
plt.title("Bakery Sales for the Week!")       #function to set title for graph
plt.xlabel("Days of the Week")                #function to set label for x axis
plt.ylabel("Sales of each day")               #function to set label for y axis

plt.show()                                    #to show the graph
```

```python
#applying some functions to plot the line chart

x = [1,2,3,4,5,6]
y = [110,100,120,145,90,125]

plt.plot(x,y,color = "red",linestyle = "--",linewidth = 1,marker = "o")
plt.title("Sales data for one week")
plt.xlabel("Days of the Week")
plt.ylabel("Sales of each day")
plt.xlim(1,7)
plt.ylim(100,150)
plt.xticks([1,2,3,4,5,6], ["Mon","Tue","Wed","Thu","Fri","Sat"])
plt.grid(color ="green",linestyle = ":",linewidth = 0.5)
plt.show()
```

```python
#some extra in line chart
x = ["Mon","Tue","Wed","Thu","Fri","Sat"]
y1 = [120,110,145,150,100,130]
y2 = [115,105,125,140,100,150]
plt.plot(x,y1,color = "red",marker = "o",label = "Sales data for week 1")
plt.plot(x,y2,color = "blue",marker = "o",label = "Sales data for week 2")
plt.title("Sales Data for 2 weeks")
plt.xlabel("Week Days")
plt.ylabel("Sales Data")
plt.legend(loc ="upper left")
plt.grid(color = "green",linestyle = ":")
plt.show()
```

#### BAR CHARTS

```python
product = ["A","B","C","D","E"]
sales = [1000,1200,800,1500,900]
plt.bar(product,sales, color = "grey",width = 0.5,label = "Sales Data")
plt.title("category wise sales data")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#use of plt.barh() --> to plot the bars horiozontally
product = ["A","B","C","D","E"]
sales = [1000,1200,800,1500,900]
plt.barh(product,sales, color = "lightgreen",label = "Sales Data")
plt.title("category wise sales data")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

#### PIE CHARTS

```python
# creating an basic pie chart
regions = ["North","South","East","West"]
revenue = [2000,1200,1500,1300]
plt.pie(revenue,labels = regions, colors = ["Gold","pink","lightgreen","yellow"],autopct = "%1.1f%%")
plt.title("Revenue generated per region")
plt.show()
```

```python
#creating a more attractive pie chart

regions = ["North","South","East","West"]
revenue = [2000,1200,1500,1300]
plt.pie(revenue,labels = regions, colors = ["Gold","pink","lightgreen","yellow"],
        autopct = "%1.1f%%",shadow = True,explode = (0.1,0,0,0))
plt.title("Revenue generated per region")
plt.show()
```

#### HISTOGRAMS

```python
scores = [83,85,95,100,55,65,57,58,68,75,73,89,84,70,57]    #marks of 15 students
plt.hist(scores, bins = 5, color = "purple", edgecolor = "black")
plt.xlabel("Scores Ranges")
plt.ylabel("Number of Students")
plt.title("Shows no. of students in each category")
plt.show()
```

#### SCATTER PLOT

```python
#this chart shows the relationship between the study hours and scores in exams
study_hours = [1,2,3,4,5,6,7,8]
scores = [60,65,70,75,80,85,90,95]

plt.scatter(study_hours,scores,color = "red",marker = "D",label = "Study Hours V/s Scores")
plt.xlabel("Study_hours")
plt.ylabel("Scores")
plt.title("Relationship between score and study hours")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# when comparining between two classes
plt.scatter([3,4,5],[52,60,80],color = "skyblue",marker = "o",label = "Class A")
plt.scatter([3,4,5],[45,57,68],color = "green",marker = "^",label = "Class B")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Compiring two classes")
plt.grid(True)
plt.legend()
plt.show()
```
