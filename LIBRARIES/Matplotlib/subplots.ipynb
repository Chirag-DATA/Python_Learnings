```python
import matplotlib.pyplot as plt
```

```python
#basic subplot
#it is not looking like a professional chart

x = [1,2,3,4]
y = [15,22,10,17]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")


plt.tight_layout()       #adjust the charts in the frame
plt.show()
```

```python
#one more basic way of subplotting

plt.figure(figsize = (12,6))
#here figsize =(9,5) represents the area of canvas that is of width(9 inches) and height(5 inches)
# if figsize is greater the charts are clearer.  

x = [1,2,3,4,5,6]
y1 = [1,4,9,16,25,36]
y2 = [1,8,9,64,125,216]

plt.subplot(2,2,1)
plt.plot(x,y1,color = "red")
plt.title("First Chart")


plt.subplot(2,2,2)
plt.plot(x,y2,color = "red")
plt.title("Second Chart")


plt.subplot(2,2,3)
plt.plot(y1,y2,color = "red")
plt.title("Third Chart")


plt.subplot(2,2,4)
plt.plot(y2,y1,color = "red")
plt.title("Fourth Chart")

plt.tight_layout()
plt.show()


```

```python
#professional way for creating subplots
#when creating subplots, always ensure that rows and columns number should equals to the charts or graphs

fig,ax = plt.subplots(1,2,figsize = (10,5))     #create the subplots 1-> rows, 3-> columns

x = [1,2,3,4]
y = [15,22,10,17]

ax[0].plot(x,y,color = "red")
ax[0].set_title("Line Chart")

ax[1].bar(x,y,color = "green")
ax[1].set_title("Bar Chart")



plt.suptitle("Comaparison of Line Chart, Bar Chart and Pie Chart")

plt.tight_layout()

plt.savefig("charts.png",dpi = 200,bbox_inches = "tight")
plt.show()

```
