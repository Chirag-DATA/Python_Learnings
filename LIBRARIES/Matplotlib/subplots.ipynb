```python
import matplotlib.pyplot as plt
```

```python
#basic subplot
#it is not looking like a professional chart

x = [1,2,3,4]
y = [15,22,10,17]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()       #adjust the charts in the frame
plt.show()
```

```python
#professional way for creating subplots
#when creating subplots, always ensure that rows and columns number should equals to the charts or graphs

fig,ax = plt.subplots(1,2,figsize = (10,5))     #create the subplots 1-> rows, 3-> columns

x = [1,2,3,4]
y = [15,22,10,17]

ax[0].plot(x,y,color = "red")
ax[0].set_title("Line Chart")

ax[1].bar(x,y,color = "green")
ax[1].set_title("Bar Chart")



plt.suptitle("Comaparison of Line Chart, Bar Chart and Pie Chart")

plt.tight_layout()

plt.savefig("charts.png",dpi = 200,bbox_inches = "tight")
plt.show()

```
