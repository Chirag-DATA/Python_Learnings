```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
#Scatter Plot
sns.scatterplot(x = "total_bill",y = "tip", data = df)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
#Line Chart
sns.lineplot(x = "size",y = "total_bill",data = df)
plt.title("Line Plot of Size vs Total Bill")
plt.show()
```

```python
#bar plot
sns.barplot(y = "total_bill",x = "day",data = df)
plt.title("Categorical Bar Plot")
plt.show()
```

```python
# Box Plot
sns.boxplot(x = 'day',y = "total_bill",data =df)
plt.title("Box Plot of Day vs Total Bill")
```

```python
# Violin Plot
sns.violinplot(x = "day", y= "total_bill",data = df)
plt.title("Violin Plot of day vs Total Bill")
```

```python
#histogram
sns.histplot(df["total_bill"],kde = True)
```

```python
#kdeplot
sns.kdeplot(df["total_bill"],fill =True)
```

```python
#Pair Plot --> scatterplots between all variable pairs
sns.pairplot(df)
```

```python
#Heat Map --> color coded matrix (correlation matrix of numerical data columns)
corr = df[['total_bill','tip','size']].corr()
print(corr)   #printing correlation

#creating map
sns.heatmap(corr, annot = True, cmap = "coolwarm")
```
